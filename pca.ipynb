```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# 📌 Cargar el dataset de ejemplo (Iris)
iris = load_iris()
X = iris.data  # Variables predictoras
y = iris.target  # Etiquetas

# 📌 Normalizar los datos (importante para PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 📌 Aplicar PCA para reducir a 2 componentes principales
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 📌 Crear un DataFrame con los componentes principales
df_pca = pd.DataFrame(X_pca, columns=["Componente 1", "Componente 2"])
df_pca["Clase"] = y

# 📌 Visualización de los datos después de PCA
plt.figure(figsize=(8,6))
scatter = plt.scatter(df_pca["Componente 1"], df_pca["Componente 2"], c=df_pca["Clase"], cmap="viridis", alpha=0.7)
plt.colorbar(scatter, label="Clases")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("PCA aplicado al dataset Iris")
plt.show()
```
