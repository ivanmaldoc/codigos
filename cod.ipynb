```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Generar datos de ejemplo
# Crear 300 puntos distribuidos en 4 clusters
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# 2. Visualizar los datos
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Datos sin agrupar")
plt.show()

# 3. Aplicar K-Means
kmeans = KMeans(n_clusters=4)  # Especificar el número de clusters
kmeans.fit(X)  # Entrenar el modelo
y_kmeans = kmeans.predict(X)  # Predecir los clusters

# 4. Visualizar los clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("Datos agrupados con K-Means")
plt.show()
```
